```python
import pandas as pd
import os
```

```python
def get_inventory_data(html_file):
    plant = pd.read_html(html_file)
    
    df_plant = plant[1]
    
    return df_plant
```

```python
path = r'%s' % os.getcwd().replace('\\','/')
html_file = path+'/library/'+'game/period3/'+'plant.html'
```

```python
df_plant = get_inventory_data(html_file)
```

```python
pd.DataFrame(df_plant[df_plant['Product'] == 'Oil'].groupby(['Company'])['Number of Plants'].sum()).to_csv('plant.csv')
```
